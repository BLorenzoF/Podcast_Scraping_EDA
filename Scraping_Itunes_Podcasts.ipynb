# Scraping Podcast information from Itunes

## Obtaining podcast links to scrape

This part of the code is taken and adapted from user Odenizgiz on [Github](https://github.com/odenizgiz/Podcasts-Data/blob/master/02.1%20Build%20the%20Dataset%20from%20iTunes%20Website.ipynb). First we scrape all the URLs from this [Itunes page](https://itunes.apple.com/us/genre/podcasts/id26?mt=2) which contains all the podcast genres. Then we scrapes each of these pages in turn to extract the links to each individual podcast page.

```python
#import libraries
import requests
import pattern3.web as web
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import re
import pickle
```

```python
webpage = 'https://itunes.apple.com/us/genre/podcasts/id26?mt=2'
```

```python
def extract_links(webpage):
    try:
        url = web.URL(webpage)
        bs = BeautifulSoup(url.download(cached = False))
        links_page = []
        for link in bs.findAll('a', href=True):
            links_page.append(link['href'])
    except:
        pass
    return links_page
```

```python
#extract all the links from the podcast genres page
links = extract_links(webpage)
```

```python
def detect_patterns(links, pattern):
    pattern_links = []
    for link in links:
            if re.search(pattern, link):
                pattern_links.append(link)
                print (link)
    return pattern_links
```

```python
pattern_genre = 'https://podcasts.apple.com/us/genre/podcasts-'
```

```python
#find only the links for the specific genre pages
genre_links = detect_patterns(links, pattern_genre)
```

```python
pattern_podcast = 'https://podcasts.apple.com/us/podcast/'
```

```python
#extract the links in each of the podcast genre pages, and from each of these
#keep only the links to podcast pages
podcast_links = []
for genre_link in genre_links:
    all_links = extract_links(genre_link)
    podcast_links.append(detect_patterns(all_links, pattern_podcast))
```

```python
podcast_links_list = list(set(np.concatenate(podcast_links).tolist()))
```

```python
#check how many links we have collected
len(podcast_links_list)
```

```python
#save the list
with open('podcast_links.pkl', 'wb') as f:
    pickle.dump(podcast_links_list, f)
```

## Extracting useful information from each link

```python
#load the list
import pickle
with open('podcast_links.pkl', 'rb') as f:
    podcast_links_list = pickle.load(f)
```

```python
def url_to_dict(url):
    import requests
    from bs4 import BeautifulSoup
    import re
    import ast
    #connect to the URL
    response = requests.get(url)
    #parse HTML and save BeautifulSoup object
    soup = BeautifulSoup(response.text, "html.parser")
    #get BS tags
    tags = soup.find_all('script')
    #get tag no. 6
    info_tag = tags[6]
    #convert tag to string
    string_text = str(info_tag)
    #remove everything up to the first bracket
    string_clean = re.sub(r'^.*?{', '{', string_text)
    #remove "script" at the end
    string_clean = string_clean.replace('</script>', '')
    #convert to dict
    podcast_dict = ast.literal_eval(string_clean)
    return podcast_dict
```

```python
#function to get information on reviews
def get_reviews (podcast_dict):
    review_id_list = []
    reviews = podcast_dict['data']['relationships']['reviews']['data']
    for item in reviews:
        review_id_list.append(item['id'])
    review_list = []
    information_list = podcast_dict['included']
    for item in information_list:
        item_type = item['type']
        item_id = item['id']
        if ((item_type == 'media/review') & (item_id in review_id_list)):
            review_list.append(item)
    review_date_list = []
    review_rating_list = []
    review_title_list  = []
    review_content_list  = []
    for item in review_list:
        try:
            review_date_list.append(item['attributes']['date'])
        except:
            pass
        try:
            review_rating_list.append(item['attributes']['rating'])
        except:
            pass
        try:
            review_title_list.append(item['attributes']['title'])
        except:
            pass
        try:
            review_content_list.append(item['attributes']['review'])
        except:
            pass
    return  review_date_list, review_rating_list, review_title_list, review_content_list, len(review_list)
```

```python
def get_episode_info (podcast_dict):
    episode_id_list = []
    episodes = podcast_dict['data']['relationships']['episodes']['data']
    for item in episodes:
        episode_id_list.append(item['id'])
    episode_list = []
    information_list = podcast_dict['included']
    for item in information_list:
        item_type = item['type']
        item_id = item['id']
        if ((item_type == 'media/podcast-episode') & (item_id in episode_id_list)):
            episode_list.append(item)
    ep_description_list = []
    ep_name_list = []
    ep_duration_list  = []
    ep_date_list = []
    for item in episode_list:
        try:
            ep_name_list.append(item['attributes']['name'])
        except:
            pass
        try:
            ep_description_list.append(item['attributes']['description']['standard'])
        except:
            pass
        try:
            ep_duration_list.append(item['attributes']['durationInMilliseconds'])
        except:
            pass
        try:
            ep_date_list.append(item['attributes']['releaseDateTime'])
        except:
            pass
    return ep_name_list, ep_description_list, ep_duration_list, ep_date_list, len(episode_list)
```

```python
#function to get podcast information
def podcast_info (podcast_dict):
    podcast_data = podcast_dict['data']
    attributes = podcast_data['attributes']
    ratings = attributes['userRating']
    #use fucntion to get information on episodes
    episode_names, episode_descriptions, episode_durations, episode_dates, num_episodes = get_episode_info(podcast_dict)
    #use function to get information on reviews
    review_dates, review_ratings, review_titles, reviews, num_reviews = get_reviews(podcast_dict)
    #build the dict with the useful information
    clean_podcast_dict = {'podcast_id' : podcast_data['id'],
                   'name' : attributes['name'],
                   'genres' : attributes['genreNames'],
                   'author' : attributes['artistName'],
                   'description' : attributes['description']['standard'],
                   'contentRating': attributes['contentRating'],
                   'avg_rating' : ratings['value'],
                   'num_ratings' : ratings['ratingCount'],
                   'num_ratings_1star' : ratings['ratingCountList'][0],
                   'num_ratings_2star' : ratings['ratingCountList'][1],
                   'num_ratings_3star' : ratings['ratingCountList'][2],
                   'num_ratings_4star' : ratings['ratingCountList'][3],
                   'num_ratings_5star' : ratings['ratingCountList'][4],
                    'episode_names' : episode_names,
                   'episode_descriptions' : episode_descriptions,
                    'episode_durations': episode_durations,
                    'episode_dates' : episode_dates,
                    'num_episodes_sample': num_episodes,
                    'review_ratings': review_ratings,
                    'review_titles' : review_titles,
                    'reviews' : reviews,
                    'review_dates' : review_dates,
                    'num_reviews_sample' : num_reviews
                    }
    return clean_podcast_dict
```

```python
#build the dataframe
podcast_info_list = []
failed_podcasts_url = []
failed_podcasts_info = []
for podcast in podcast_links_list:
    try:
        long_dict = url_to_dict(podcast)
    except:
        failed_podcasts_url.append(podcast)
    try:
        short_dict = podcast_info(long_dict)
        podcast_info_list.append(short_dict)
    except:
        failed_podcasts_info.append(podcast)
df = pd.DataFrame(podcast_info_list)
```
